```python
import findspark
findspark.init()

from pyspark.sql.session import SparkSession
from pyspark.sql.functions import col
from pyspark.sql import SQLContext
```

```python
spark = SparkSession \
    .builder \
    .appName("Python Spark SQL basic example") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()

df = spark.read.option(
    "multiline", "true"
).option(
    "quote", '"'
).option(
    "escape", '"'
).csv(
    "./../datasets/netflix_reviews.csv", 
    header=True, 
    sep=","
)
```

```python
cols = ["userName", "score"]
```

```python
print(df.select(*cols).filter(df.score.isNotNull()).orderBy(df.score.desc()).show())
```

```python
fd = df.select(*cols).filter(df.score.isNotNull())
fd.sort(col("score")).collect()

fd.collect()
```
